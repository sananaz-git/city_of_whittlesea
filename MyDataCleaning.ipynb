#For Q9. How long has the person lived at this address?

```python
person_df.Q9.value_counts()
```

```python
person_df.Q9.isna().sum() # 26 NAs
```

Following values will be assigned to each duration of stay:

| Length of stay | Scale |
| ----------- | -------- |
| Less than 1 year | 0.5
| 1 to 5 years | 1.5
| 5 to 10 years | 7.5
| 10 years or more | 15
| Nulls | median of the column

```python
# Generating conditions for each duration of stay and values to be assigned
conditions = [person_df['Q9'].eq(1.0), person_df['Q9'].eq(2.0), person_df['Q9'].eq(3.0), person_df['Q9'].eq(4.0)]
values = [0.5,1.5,7.5,15]
```

```python
# Substituting the nan values with median
person_df['Q9'] = np.select(conditions, values, default=person_df['Q9'].median())
```

Verifying that the values get assigned properly as in the 4th and 5th last records there were nan values. We can see 3 is the median value.

```python
# Creating a list matches that contains match of each Q27 entry against the industry list created above.
matches=[]

for x in person_df.Q27:
    try:
        matches.append([process.extract(x, industry_list, limit=1)][0][0][0])
    except:
        matches.append(np.nan)
```
